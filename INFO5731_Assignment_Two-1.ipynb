(3) Collect all the reviews of the top 100 most popular software from [G2](https://www.g2.com/) or [Capterra](https://www.capterra.com/)

(4) Collect the abstracts of the top 10000 research papers by using the query [natural language processing](https://citeseerx.ist.psu.edu/search?q=natural+language+processing&submit.x=0&submit.y=0&sort=rlv&t=doc) from CiteSeerX.

```python
#importing required modules
import requests  
from bs4 import BeautifulSoup as bs 
import re


```

```python
#opening empty list
iphone_reviews=[]
#running loop for 10000 pages
for i in range(1,10000):
  ip=[]  
  url="https://www.amazon.com/Apple-iPhone-11-64GB-Unlocked/product-reviews/B07ZPKF8RG/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"+str(i)
  response = requests.get(url)
  #using bs to extract HTML content
  soup = bs(response.content,"html.parser") 
  #extracting review text using their class tags
  reviews = soup.find_all("div",attrs={"class","a-row a-spacing-small review-data"})  
  for i in range(len(reviews)):
    ip.append((reviews[i].text).replace('\n',''))
    iphone_reviews=iphone_reviews+ip
```

```python
#importing list into df
import pandas as pd
reviews= pd.DataFrame(iphone_reviews,columns=['reviews'])
reviews
```

```python
reviews.to_csv('data.csv', index=False)
```

# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# importing text into a df
import pandas as pd
data_url="https://raw.githubusercontent.com/nivemula/Nithin_INFO5731_Spring2022/main/textdata.txt"
df = pd.read_table(data_url,names=['text'])
df

```

```python
#Number of characters
df['Number of Characters'] = df['text'].str.len()
df
#import re

#df['Reviews after Noise Removal'] = df['Review'].str.replace('[^\w\s]','')
#df['Reviews after Noise Removal'] = df['Reviews after Noise Removal'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x )
```

```python
#Number of numerics
df['Numbers of numerics'] = df['text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

```python
#digit removal
df['After digits removal'] = df['text'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df
```

```python
#Stopwords removal
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stopwords Removal'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

```python
#Lower casing
df['Lower Case'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```python
#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['After Stemming'] = df['Lower Case'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

```python
#Punctuation removal
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

```python
#Lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')
df['After Lemmatization'] = df['Removal of Punctuation'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```python
df.to_csv('cleaned_data.csv',index=False)
```

```python
#(1) Parts of Speach Tagging
#importing and downloading required modules
import nltk
nltk.download('averaged_perceptron_tagger')
from nltk.tokenize import word_tokenize
#opening a empty list
pos = []
#tokenising each word from sentences
for sentence in df['Lower Case']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos

```

```python
# Parts of Speech Count
from collections import Counter
counts = []
#counting each word and token
for tags in pos:
  count = Counter( tag for word,  tag in tags)
  counts.append(count)
counts
```
